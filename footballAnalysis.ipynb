```python

player_salary=player_salary.sort_values(by='wage_eur',ascending=False)
```

```python
#plot names vs salary
plt.figure(figsize=(8,5))
plt.bar(list(player_salary['short_name'])[0:5],list(player_salary['wage_eur'])[0:5],color=["violet","blue","green","yellow","red"])
plt.show()
```
